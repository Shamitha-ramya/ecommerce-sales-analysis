```python
import pandas as pd
```

```python
order_items=pd.read_csv(
    '/Users/shamitharamya/Documents/ecommerce-sales-analysis/data/raw_data/olist_order_items_dataset.csv',
    dtype={'price': 'float', 'freight_value': 'float'},
    parse_dates=['shipping_limit_date']
)
```

```python
orders=pd.read_csv('/Users/shamitharamya/Documents/ecommerce-sales-analysis/data/raw_data/olist_orders_dataset.csv',
                   parse_dates=['order_purchase_timestamp','order_approved_at','order_delivered_carrier_date','order_delivered_customer_date','order_estimated_delivery_date'])
```

```python
order_items.shape
```

```python
revenue_data=pd.merge(orders[['order_id','order_status','order_purchase_timestamp']],order_items[['order_id','product_id','price','freight_value']],on='order_id')
```

```python
revenue_data
```

```python
revenue_data['year']=revenue_data['order_purchase_timestamp'].dt.year
```

```python
revenue_data=revenue_data[revenue_data['order_status']=='delivered']
```

```python
revenue_data['revenue']=revenue_data['price']+revenue_data['freight_value']
```

```python
#year wise revenue
revenue_data.groupby('year')['revenue'].sum()
```

```python
#which year has highest revenue
revenue_data.groupby('year')['revenue'].sum().idxmax()
```

```python
revenue_data['month']=revenue_data['order_purchase_timestamp'].dt.month
```

```python
#month wise revenue
revenue_data.groupby('month')['revenue'].sum()
```

```python
#which month has highest revenue
revenue_data.groupby('month')['revenue'].sum().idxmax()
```

```python
#product wise revenue
revenue_data.groupby('product_id')['revenue'].sum()
```

```python
#which product has highest revenue
revenue_data.groupby('product_id')['revenue'].sum().idxmax()
```
