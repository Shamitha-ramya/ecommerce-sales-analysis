```python
orders_customers['customer_state'].value_counts().head()
```

'SP' is the state with most orders

```python
states = orders_customers['customer_state'].value_counts()
```

```python
orders_customers.groupby('customer_state')['customer_city'].nunique()
```

```python
weighted_states = orders_customers.groupby('customer_state')['customer_city'].nunique()
```

```python
orders_per_100_cities = (states / weighted_states) * 100
```

This metric normalizes order volume by the number of cities in each state, allowing a fair comparison across states with different geographic distributions.

Interpretation

States such as DF, RJ, and SP exhibit high order density, indicating stronger e-commerce market penetration and higher purchasing activity per city.

In contrast, some states with a larger number of cities show lower order density, suggesting comparatively lower purchasing activity per city. This may indicate:

lower market penetration

lower purchasing power

logistical or accessibility differences in e-commerce adoption.
